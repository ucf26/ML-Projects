# Model Training & Prediction

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, random_state=42, init='k-means++')
```

```python
kmeans.fit(X)
```

```python
kmeans.n_clusters
```

```python
kmeans.cluster_centers_
```

```python
y = kmeans.labels_
y
```

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1])
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
```

# Finding The Optimal Number of Clusters 

## Elbow Method
The elbow method is a technique used in cluster analysis to determine the optimal number of clusters (k) for a given dataset. It helps to identify the point where adding more clusters does not significantly improve the model's performance, which is often visualized as an "elbow" in a plot.

### Steps to Implement the Elbow Method:

1. **Choose a Range of k Values**: Decide on a range of values for k (e.g., from 1 to 10).

2. **Calculate the Within-Cluster Sum of Squares (WCSS)**: For each value of k, fit a clustering algorithm (like K-Means) and compute the WCSS, which measures the compactness of the clusters.

3. **Plot the WCSS Against k**: Create a plot with the number of clusters on the x-axis and the WCSS on the y-axis.

4. **Identify the Elbow Point**: Look for the point where the rate of decrease in WCSS sharply changes, indicating that adding more clusters does not provide significantease let me know!

```python
wcss = [] # within cluster some square
for i in range(1, 11):
    tmp_kmeans = KMeans(n_clusters=i, random_state=42, init='k-means++')
    tmp_kmeans.fit(X)
    wcss.append(tmp_kmeans.inertia_)

wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
```
